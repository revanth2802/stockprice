```python
if 'Date' in df.columns:
    # Splitting the 'Date' column to extract year, month, and day
    splitted = df['Date'].str.split('-', expand=True)

    # Check if the split operation created expected columns
    if len(splitted.columns) == 3:
        # Creating new columns for year, month, and day
        df['year'] = splitted[0].astype('int')
        df['month'] = splitted[1].astype('int')
        df['day'] = splitted[2].astype('int')

        # Displaying the updated DataFrame
        print(df.head())
    else:
        print("Error: Unexpected number of columns after splitting 'Date'.")
else:
    print("The 'Date' column is not found in the DataFrame.")
```

```python
df['Date'] = pd.to_datetime(df['Date'])

# Create the 'is_quarter_end' column
df['is_quarter_end'] = df['Date'].dt.is_quarter_end

# Check the DataFrame to ensure the new column is added correctly
print(df[['Date', 'is_quarter_end']].head())

# Group by 'is_quarter_end' and calculate the mean
data_grouped = df.groupby('is_quarter_end').mean()

# Display the grouped data
print(data_grouped)
```

```python
df['open-close'] = df['Open'] - df['Close']
df['low-high'] = df['Low'] - df['High']
df['target'] = np.where(df['Close'].shift(-1) > df['Close'], 1, 0)
```

```python
plt.pie(df['target'].value_counts().values,
		labels=[0, 1], autopct='%1.1f%%')
plt.show()
```

```python
plt.figure(figsize=(10, 10))

# As our concern is with the highly
# correlated features only so, we will visualize
# our heatmap as per that criteria only.
sb.heatmap(df.corr() > 0.9, annot=True, cbar=False)
plt.show()
```

```python
features = df[['open-close', 'low-high', 'is_quarter_end']]
target = df['target']

scaler = StandardScaler()
features = scaler.fit_transform(features)

X_train, X_valid, Y_train, Y_valid = train_test_split(
	features, target, test_size=0.1, random_state=2022)
print(X_train.shape, X_valid.shape)
```

```python
models = [LogisticRegression(), SVC(
kernel='poly', probability=True), XGBClassifier()]

for i in range(3):
  models[i].fit(X_train, Y_train)

print(f'{models[i]} : ')
print('Training Accuracy : ', metrics.roc_auc_score(
	Y_train, models[i].predict_proba(X_train)[:,1]))
print('Validation Accuracy : ', metrics.roc_auc_score(
	Y_valid, models[i].predict_proba(X_valid)[:,1]))
print()
```
